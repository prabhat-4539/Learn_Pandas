•	df.isnull().sum() → Count missing values in each column.
NOTE: df.isna() and df.isnull() are same. Both will return a dataframe in which only NA or null values will be marked as True, all other will be marked as False.

```python
ufo.isna()
```

```python
ufo.isnull()
```

```python
sum(ufo.isna()['Shape Reported'])
#Note: Here the built-in python sum function sums all the value of the pandas series. (ufo.isna()['Shape Reported'] is a pandas series of booleans) 
```

```python
ufo.isna().sum()
# Here the sum method of pandas dataframe sums all the values of the colmns of the dataframe. (ufo.isna() is a dataframe of booleans)
```

sum() exists both as a method of a Pandas Series and as a built-in Python function, but they behave slightly differently.
