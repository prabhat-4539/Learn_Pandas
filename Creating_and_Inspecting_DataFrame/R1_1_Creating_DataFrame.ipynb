# CREATING DATAFRAME

DataFrame is not a method of Pandas; it is a class in the pandas module.
A DataFrame in Pandas is a two-dimensional, tabular data structure similar to an Excel spreadsheet or SQL table. It consists of labeled rows and columns.
To create a dataframe.

```python
# Method1 : directly passing a dictionary in dataframe class.
student_details = pd.DataFrame({'Roll_num':[1,2,3,4],'Name':['Prabhat','Sarika','Ashwani','Shubham'],'Age':[27,27,25,23]})
print(student_details)
```

```python
# Method 2 : making a dictionary first and then passing it into dataframe class.
student_dict = {'Roll_num':[1,2,3,4],'Name':['Prabhat','Sarika','Ashwani','Shubham'],'Age':[27,27,25,23]}
student_details_df = pd.DataFrame(student_dict)
print(student_details_df)
```

```python
# Method 3 : passing a list of list and then using column parameter. Here each list is one row.
student_details_table = pd.DataFrame([[1,'Prabhat',27],[2,'Sarika',27],[3,'Ashwani',25],[4,'Shubham',23]],columns=['Roll_no','Name','Age'])
print(student_details_table)
```
